```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the CSV file
csv_data = pd.read_csv('Desalination_Combined_1Hz_19-04-2024.csv')

# Plot the graph 'Id vs MRU_Surge_m'
plt.figure(figsize=(10, 6))
plt.plot(csv_data['Id'], csv_data['MRU_Surge_m'], color='blue', label='MRU_Surge_m')
plt.title('Id vs MRU_Surge_m')
plt.xlabel('Id')
plt.ylabel('MRU_Surge_m (m)')
plt.legend()
plt.grid()
plt.show()
   
```

```python
# Plot the graph 'Time vs MRU_Surge_m'
plt.figure(figsize=(10, 6))
plt.plot(csv_data['Time'], csv_data['MRU_Surge_m'], color='blue', label='MRU_Surge_m')
plt.title('Time vs MRU_Surge_m')
plt.xlabel('Time')
plt.ylabel('MRU_Surge_m (m)')
plt.legend()
plt.grid()
plt.show()
```
